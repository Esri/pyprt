TASK: Implement simple application that does repeated generation of an initial shape with varying attributes, and returns a report, on which it does some simple numerical processing.

```python
import sys
import os

import pyprt
from pyprt.pyprt_utils import visualize_prt_results

import pandas as pd
```

```python
CS_FOLDER = os.getcwd()
def asset_file(filename):
    return os.path.join(CS_FOLDER, '..', 'data', filename)
```

#### PRT Initialization

```python
print("\nInitializing PRT.")
pyprt.initialize_prt()

if not pyprt.is_prt_initialized():
    raise Exception("PRT is not initialized")
```

```python
v = [0, 0, 0,  0, 0, 20,  10, 0, 10,  10, 0, 0]
initialGeometry = pyprt.InitialShape(v)
rpk = asset_file("simple_rule2019.rpk")
attrs = {'ruleFile': 'bin/simple_rule2019.cgb', 'startRule': 'Default$Footprint'}

mod = pyprt.ModelGenerator([initialGeometry])
generated_model = mod.generate_model([attrs], rpk, 'com.esri.pyprt.PyEncoder', {})

visualize_prt_results(generated_model)
```

```python
initialGeometry2 = pyprt.InitialShape([0, 0, 0,  0, 0, 10,  10, 0, 10,  10, 0, 0])
initialGeometry3 = pyprt.InitialShape([0, 0, 0,  0, 0, 30,  10, 0, 10,  10, 0, 0])

rpk3 = asset_file("simpleRuleDataset.rpk")
attrs3 = {'ruleFile': 'bin/simpleRuleDataset.cgb', 'startRule': 'Default$Lot'}
```

```python
reports = []
model_to_generate = pyprt.ModelGenerator([initialGeometry, initialGeometry2, initialGeometry3])
model_test = model_to_generate.generate_model([attrs3], rpk3, 'com.esri.pyprt.PyEncoder', {'emitGeometry': False})
for val in range(0,10):
    
    attrs3['maxHeight'] = float(val)
    models = model_to_generate.generate_model([attrs3])

    for model in models:
        if model:
            reports.append(model.get_report())
```
